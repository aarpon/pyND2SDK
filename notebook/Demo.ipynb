# pyND2SDK demo

```python
# Import the reader
from pyND2SDK.nd2Reader import nd2Reader
```

```python
# Example file
filename = "./file01.nd2"
```

```python
# Instantiate the reader
r = nd2Reader()
```

```python
# Open the file
r.open(filename)
```

```python
?r.get_geometry()
```

```python
# Get geometry information
r.get_geometry()
```

```python
# Get attributes
r.get_attributes()
```

```python
# Get first Picture
p = r.load_by_index(0)
```

```python
# Summary of the Picture
type(p), p
```

```python
# Get the first component (channel)
img = p[0]
```

```python
type(img)
```

```python
import matplotlib.pyplot as plt
```

```python
# Show the picture
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Load the picture by substr
?r.load
```

```python
# We retrieve the same picture as before but by specifying (almost) all indices
q = r.load(time = 0, position = 0, plane = 0)
```

```python
# Work on the image
from scipy.ndimage.filters import gaussian_filter
f_q = blurred = gaussian_filter(q[0], sigma=15)
```

```python
# Show the picture
plt.imshow(f_q, cmap='gray')
plt.show()
```
